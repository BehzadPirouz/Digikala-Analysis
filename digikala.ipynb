## Step 2: Data Cleaning

```python
df_orders_clean = df_orders.copy()
df_orders_clean["DateTime_CartFinalize"] = pd.to_datetime(df_orders_clean["DateTime_CartFinalize"])
df_orders_clean = df_orders_clean[df_orders_clean["Amount_Gross_Order"] > 0]
df_orders_clean
```

```python
df_product_clean = df_product[["id", "product_title_fa", "category_title_fa", "brand_name_fa"]].copy()
df_product_clean.rename(columns={"id": "product_id"}, inplace=True)
df_product_clean
```

```python
df_tarikh_clean = df_tarikh[["product_id", "selling_price", "created_at"]].copy()
df_tarikh_clean["selling_price"] = pd.to_numeric(df_tarikh_clean["selling_price"], errors="coerce")
df_tarikh_clean = df_tarikh_clean.dropna()
df_tarikh_clean
```

### Merge Tables

```python
df_merge = pd.merge(df_tarikh_clean, df_product_clean, on="product_id", how="inner")
df_final = pd.merge(df_merge, df_orders_clean, left_on="product_id", right_on="ID_Item", how="inner")
df_final.head()
# df_final.info()
# df_final.isnull().sum()
```

## Step 3: Process

```python
city_product_sales = df_final.groupby(["city_name_fa", "product_id", "product_title_fa"]).agg({"Quantity_item": "sum"}).reset_index()
city_product_sales
```

```python
city_product_sales["Quantity_item"] = pd.to_numeric(city_product_sales["Quantity_item"], errors="coerce")
city_product_sales = city_product_sales.dropna(subset=["Quantity_item"])
top_bottom_products = []
for city in city_product_sales["city_name_fa"].unique():
    city_data = city_product_sales[city_product_sales["city_name_fa"] == city]
    if len(city_data) < 3: continue
    top3 = city_data.nlargest(3, "Quantity_item").copy()
    top3["Status"] = "Top"
    bottom3 = city_data.nsmallest(3, "Quantity_item").copy()
    bottom3["Status"] = "Bottom"
    combined = pd.concat([top3, bottom3])
    top_bottom_products.append(combined)
result_df = pd.concat(top_bottom_products).reset_index(drop= True)
result_df
```

## Step 4: Visualize

```python
df_final["Season_Number"] = ((df_final["DateTime_CartFinalize"].dt.month - 3) % 12) // 3 + 1
season_map = {1: "Spring", 2: "Summer", 3: "Autumn", 4: "Winter"}
df_final["Season"] = df_final["Season_Number"].map(season_map)
df_final[["DateTime_CartFinalize", "Season_Number", "Season"]]
```

```python
df_final.to_csv(r"C:\Users\Behzad\Documents\project\df_final.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
df_final["total_sales"] = df_final["Quantity_item"] * df_final["selling_price"]
seasonal_sales = df_final.groupby(["city_name_fa", "Season"])["total_sales"].sum().reset_index()
top_cities = df_final.groupby("city_name_fa")["total_sales"].sum().nlargest(3).index.tolist()
seasonal_top = seasonal_sales[seasonal_sales["city_name_fa"].isin(top_cities)]
plt.figure(figsize= (10, 6))
sb.barplot(data= seasonal_top, x= "Season", y= "total_sales", hue= "city_name_fa")
plt.title("seasonal sales for top 3 profitable cities")
plt.xlabel("Season")
plt.ylabel("Total Sales")
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig("seasonal_top_cities.png")
plt.show()
```

```python
import matplotlib.ticker as ticker
df_tehran_spring = df_final[(df_final["city_name_fa"] == "تهران") & (df_final["Season_Number"] == 1)].copy()
df_tehran_spring.loc[:, "Total"] = df_tehran_spring["selling_price"] * df_tehran_spring["Quantity_item"]
daily_sales = df_tehran_spring.groupby(df_tehran_spring["DateTime_CartFinalize"].dt.date)["Total"].sum().reset_index()
q1 = daily_sales["Total"].quantile(0.25)
q3 = daily_sales["Total"].quantile(0.75)
qr = q3 - q1
upper_bound = q3 + 1.5 * qr
daily_sales_filtered = daily_sales[daily_sales["Total"] <= upper_bound].copy()
daily_sales_filtered["Moving_Avg"] = daily_sales_filtered["Total"].rolling(7, min_periods=1).mean()
plt.Figure(figsize= (14, 6))
plt.plot(daily_sales_filtered["DateTime_CartFinalize"], daily_sales_filtered["Total"], marker= "o", color= "hotpink", label= "Daily Sales")
plt.plot(daily_sales_filtered["DateTime_CartFinalize"], daily_sales_filtered["Moving_Avg"], color= "purple", linewidth=2, label= "7-Day Moving Avg")

plt.title("Tehran Daily Sales in Spring", fontsize= 14)
plt.xlabel("Date", fontsize= 12)
plt.ylabel("Sales", fontsize= 12)
plt.xticks(rotation= 45)
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("{x: .0f}"))
plt.grid(True)
plt.tight_layout()
plt.savefig("tehran_spring.png")
plt.show()
```
